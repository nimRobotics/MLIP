### Case I: MNIST dataset

For both train as well as test data the first digit in the ith row this the label (a number in the range 0-9) of the i sample. The next 784 (28x28) digits in the same row are the values of the pixels of the sample (image).

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
import cv2
from collections import Counter

# load train data
train = loadmat('mnist_train.mat')
train=train['traindata']
print("MNIST train dataset size:",train.shape)

# load test data
test = loadmat('mnist_test.mat')
test=test['testdata']
print("MNIST test dataset size:",test.shape)

print("Visualizing the MNIST dataset")
fig=plt.figure(figsize=(8, 8))
columns = 5
rows = 5
for i in range(1, columns*rows +1):
    img=train[i][1:].reshape((28, 28))
    fig.add_subplot(rows, columns, i)
    plt.imshow(img,cmap='gray')
plt.show()
```

```python
# returns euclidean_distance vector
def euclidean_distance(test,train_feature):
    # print(train_feature.shape[0])
    distance = np.zeros((5000, 1))
    for i in range(train_feature.shape[0]):
        distance[i] = np.linalg.norm(train_feature[i]-test)
    # print(distance)
    return distance

# return histogram feature from the images
def features(test,train):
    # extract feature from traindata
    train_feature = np.zeros((5000, 256))
    for j in range(train.shape[0]):
        # feature = features(train[j][1:])
        feature = cv2.calcHist([train[j][1:]], [0], None, [256], [0, 256])
        train_feature[j]=feature.reshape((1,256))

    # extract feature from traindata
    test_feature = np.zeros((500, 256))
    for j in range(test.shape[0]):
        # feature = features(test[j][1:])
        feature = cv2.calcHist([test[j][1:]], [0], None, [256], [0, 256])
        test_feature[j]=feature.reshape((1,256))
    return train_feature,test_feature

# Calculate accuracy percentage between two lists
def accuracy_metric(actual, predicted):
	correct = 0
	for i in range(len(actual)):
		if actual[i] == predicted[i]:
			correct += 1
	return correct / float(actual.shape[0]) * 100.0
```

```python
def knn(test,train,k):
    train_labels = train[:,[0]]
    test_labels = test[:,[0]]

    # get feature vectors using histogram
    # train_feature,test_feature=features(test,train)

    # get feature vectors as image pixels
    train_feature = train[:,1:]
    test_feature = test[:,1:]

    predict=np.zeros((500, 1))
    # loop over the test data
    for i in range(test_feature.shape[0]):
        distances = euclidean_distance(test_feature[i],train_feature)
        # append the training label as column
        dist = np.append(distances, train_labels, axis = 1)
        # sort based on first col i.e. distances and select knn
        neighbours = dist[dist[:,0].argsort()][0:k]
        # print(neighbours)
        class_counter = Counter()
        # voting
        for neighbor in neighbours:
            class_counter[neighbor[1]] += 1
        # print(class_counter.most_common(1)[0][0])
        predict[i]=class_counter.most_common(1)[0][0]
    accuracy=accuracy_metric(test_labels,predict)
    print("Accuracy for k = ",k," : ",accuracy)
```

```python
for i in [1,3,5,9,11]:
    accuracy=knn(test,train,i)
```
