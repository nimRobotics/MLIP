Method used for multi-class bayesian classification: **One-vs-All Classification** (Build N different binary classifiers. For the ith classifier, let the positive examples be all the points in class i, and let the negative examples be all the points not in class i.)

In the given case we have 3 classes, so we have 3 ($N\choose k$) binary classifications decision boundaries. Hence using the **One-vs-All Classification** we will have to perform binary classification on two classes w1 and w2.
The decision boundary can be found using

$$ P(\omega_1)P(x/\omega_1)=P(\omega_2)P(x/\omega_2) $$
$$ P(\omega_1)\frac{1}{\sqrt{2\pi}\sum^{1/2}_1}e^{\frac{-1}{2}(x-\mu_1)^T\sum^{-1}_1(x-\mu_1)} =     P(\omega_2)\frac{1}{\sqrt{2\pi}\sum^{1/2}_2}e^{\frac{-1}{2}(x-\mu_2)^T\sum^{-1}_2(x-\mu_2)} $$

Decision rule:
$$ x \in w1 \text{ if } (x-\mu_1)^T\sum^{-1}_1(x-\mu_1) - (x-\mu_2)^T\sum^{-1}_2(x-\mu_2) + \ln(\sum_1) - \ln(\sum_2) \text{ else } x \in w2$$
